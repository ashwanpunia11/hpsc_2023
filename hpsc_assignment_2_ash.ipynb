```python
import matplotlib.pyplot as plt
import numpy as np
import timeit
toll_limit=0.1;
```

```python
def Ash_exp(x, n):
    result = 1.0;
    term=1;
    for i in range(1,n+1):
        term = term*x/i
        result +=term
    return result

```

```python
#Plot the exp graph using inbuilt function
x = np.linspace(0,100,1000)
y = np.exp(x)

plt.figure()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.legend(['e^x with my function'])
```

```python
#Plot the exp graph using inbuilt function
x1 = np.linspace(0,100,1000)
y1 = Ash_exp(x1,100)


plt.figure()
plt.plot(x1, y1,'r')
plt.xlabel('x1')
plt.ylabel('Ash_exp(x)') 
plt.legend(['e^x with inbuilt numpy function'])
```

```python
plt.plot(x, y,'r',label='Inbuil Function')
plt.plot(x1, y1,'*',label='my Function')
plt.legend()
plt.show()
```

```python

t1=timeit.timeit(lambda: Ash_exp(5, 100));
t2=timeit.timeit(lambda: np.exp(5));

print(f"The executation time for user defined function = {t1}");
print(f"The executation time for inbuilt function = {t2}");
print(f"The difference of executation time of Ash_exp() and numpy.exp() is = {t1-t2}")

```
